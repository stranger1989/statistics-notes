### factorial

$$n!=n \times (n-1) \times (n-2) \times \dots \times 2 \times 1$$


```python
print(math.factorial(5))
print(math.factorial(0))
```

### permutation

$${}_n \mathrm{P}_{r}= n \times (n-1) \times (n-2) \times \dots \times (n-r+1)$$


```python
math.perm(5, 3)
```

### combination

$${}_n \mathrm{C}_{r}= \displaystyle \frac{{}_n \mathrm{P}_{r}}{r!} =\displaystyle \frac{n!}{(n-r)! r!}$$


```python
math.comb(5, 3)
```

### events

$$\Omega$$


```python
omega = set([1, 2, 3, 4, 5, 6, 7])
setA = set([1, 2, 3, 4])
setB = set([3, 4, 5, 6])
```

### complementary event

$$A^c$$


```python
omega.difference(setA)
```

### empty set

$$\phi$$


```python
set()
```

### union of events

$$A\cup B$$


```python
setA.union(setB)
```

### intersection

$$A\cap B$$


```python
setA.intersection(setB)
```

### venn


```python
venn2([setA, setB], ("Event A", "Event B"))
```

### probability

$$P(A)=\displaystyle \frac{k}{n}$$


```python
venn2([omega, setA], ("Event Omega", "Event A"))
```

```python
k = len(setA)
n = len(omega)
P_A = k / n
P_A
```

### conditional probability

$$P(A|B) = \displaystyle \frac{P(A \cap B)}{P(B)}$$


```python
P_AB = len(setA.intersection(setB)) / len(setB)
P_AB
```

### Bayes' theorem

$$P(B|A) = \frac{P(B)\cdot P(A|B)}{P(A)}$$


```python
P_BA = len(setB) * P_AB / len(setA)
P_BA
```
