## probability variable and probability distribution

### probability distribution


#### discrete & probability mass function

$$f(x)=P(X=x)$$


```python
stats.randint(1, 7).rvs()
```

```python
x = np.arange(1, 7)
y = stats.randint(1, 7).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

#### continuous & probability density function

$$f(x)=f(X=x)$$
$$f(x) \geq 0, \hspace{20px} \int_{-{\infty}}^{\infty} f(x)dx = 1$$
$$P(a \leq X \leq b)= \displaystyle \int_a^b f(x)dx$$


```python
stats.uniform(0, 1).rvs()
```

```python
x = np.linspace(-1, 2, 100)
y = stats.uniform(loc=0, scale=1).pdf(x)
plt.plot(x, y)
```

#### cumulative distribution function

$$F(x)=P(X \leq x)= \displaystyle \sum_{X\leq x} P(X)$$
$$F(x)=P(X \leq x)=\displaystyle \int_{- \infty}^{x} f(t)dt$$


```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
plt.plot(x, y)
```

```python
y = stats.norm.cdf(x)
plt.plot(x, y)
```

### random variable `expected value` and `variance`


#### discrete

$$E(X)=\displaystyle \sum_{i=1}^n x_i \times p_i$$


```python
X = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]
E_X = sum(x * p for x, p in zip(X, probabilities))
E_X
```

$$V(X)= \displaystyle \sum_{i=1}^n \left( x_i -\mu \right)^2 p_i$$


```python
V_X = sum((x - E_X) ** 2 * p for x, p in zip(X, probabilities))
V_X
```

#### continuous

$$E(X)= \displaystyle \int_{-\infty}^{\infty} xf(x)dx$$


```python
E_X = stats.norm(5, 2).mean()
E_X
```

$$V(X)= \displaystyle \int_{-\infty}^{\infty} (x-\mu )^2 f(x)dx$$


```python
V_X = stats.norm(5, 2).var()
V_X
```

### discrete random variable various patterns


#### bernoulli

$$P(X=1)=p$$
$$P(X=0)=1-p$$


```python
x = np.arange(-1, 3)
y = stats.bernoulli(0.5).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

$$\begin{equation*} E[X]=p \end{equation*}$$


```python
stats.bernoulli(0.5).mean()
```

$$\begin{equation*} V[X]=p(1-p) \end{equation*}$$


```python
stats.bernoulli(0.5).var()
```

#### binomial

$$P(X = x) = {}_n\mathrm{C}_xp^x(1 - p)^{n - x}$$


```python
n = 10
p = 0.5
x = np.arange(0, 11)
y = stats.binom(n, p).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

$$E(X)=np$$


```python
stats.binom(n, p).mean()
```

$$V(X)=np(1-p)$$


```python
stats.binom(n, p).var()
```

#### poisson

$$P(X=k)= \displaystyle\frac{e^{-\lambda} \lambda^{k}}{k!}$$


```python
p_lambda = 0.1
x = np.arange(0, 11)
y = stats.poisson(p_lambda).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

$$E(X)=\lambda$$


```python
stats.poisson(p_lambda).mean()
```

$$V(X)=\lambda$$


```python
stats.poisson(p_lambda).var()
```

#### geometric distribution

$$P(X=k)=(1-p)^{k-1}  p (k=1,2,3,\cdots)$$


```python
p = 0.1
x = np.arange(1, 11)
y = stats.geom(p).pmf(x)
plt.plot(x, y, "o")
plt.vlines(x, 0, y)
```

$$\displaystyle E(X)=\frac{1}{p}$$


```python
stats.geom(p).mean()
```

$$\displaystyle V(X)=\frac{1-p}{p^{2}}$$


```python
stats.geom(p).var()
```

### continuous random variable various patterns


#### normal distribution

$$f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp \left(-\frac{(x - \mu)^2}{2\sigma^2} \right) \hspace{20px} (-\infty < x < \infty)$$


```python
x = np.linspace(-3, 3, 100)
y = stats.norm().pdf(x)
plt.plot(x, y)
```

$$E(X)=\mu$$


```python
stats.norm().mean()
```

$$V(X)=\sigma^{2}$$


```python
stats.norm().var()
```

#### exponential distribution

$$f(x) = \begin{cases} \lambda e^{- \lambda x} & x \geq  0 \\ 0 & x < 0\end{cases}$$


```python
x = np.linspace(0, 1, 100)
y = stats.expon().pdf(x)
plt.plot(x, y)
```

$$\displaystyle E(X)= \frac{1}{ \lambda }$$


```python
stats.expon().mean()
```

$$\displaystyle V(X)= \frac{1}{ \lambda ^{2} }$$


```python
stats.expon().var()
```

#### continuous uniform distribution

$$\displaystyle f(x)=\frac {1}{b-a}~~~(a \leq X \leq b)$$
$$\displaystyle f(x)=0~~~(X < a, X > b)$$


```python
x = np.linspace(-1, 2, 100)
y = stats.uniform().pdf(x)
plt.plot(x, y)
```

$$\displaystyle E(X)=\frac {a+b}{2}$$


```python
stats.uniform().mean()
```

$$\displaystyle V(X)=\frac{(b-a)^{2}}{12}$$


```python
stats.uniform().var()
```
