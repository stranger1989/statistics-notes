## time series


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
# https://fred.stlouisfed.org/series/M1109BUSM293NNBR
df = sns.load_dataset("dowjones")
df = df.set_index("Date")
df.head()
```

```python
df.plot()
```

### time diff


```python
df.diff().head()
```

```python
df.diff().plot()
```

### time ratio


```python
(df / df.shift()).head()
```

```python
(df / df.shift()).plot()
```

### time pct change


```python
df.pct_change().head()
```

```python
df.pct_change().plot()
```

### time index ratio


```python
(df / df["Price"][0] * 100).head()
```

```python
(df / df["Price"][0] * 100).plot()
```
